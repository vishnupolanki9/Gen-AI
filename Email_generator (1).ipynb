```python


!pip install -q google-genai

from google.colab import userdata
from google import genai
from datetime import datetime


# Load API Key from Colab Secrets

API_KEY = userdata.get("GOOGLE_API_KEY")

client = genai.Client(api_key=API_KEY)


# Email Generator Function

def generate_email(recipient, purpose, tone, sender_name):

    today = datetime.now().strftime("%d %B %Y")

    system_instruction = """
You are a professional email writer.

Generate a clear, polite, and well-structured email.

The email must include:
1. Current Date
2. Subject Line
3. Greeting
4. Professional Email Body
5. Professional Closing
6. Sender Name

Return only the email without any extra explanation.
"""

    prompt = f"""
Date: {today}

Recipient: {recipient}

Purpose: {purpose}

Tone: {tone}

Sender Name: {sender_name}

Write a professional email.

Requirements:
- Include today's date at the top.
- Create a suitable subject line.
- Start with an appropriate greeting.
- Write a professional email body.
- End with a professional closing.
- Sign the email using the sender name.
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
        config={
            "system_instruction": system_instruction,
            "temperature": 0.3
        }
    )

    email = response.text

    # Save email into text file
    with open("email_output.txt", "w", encoding="utf-8") as file:
        file.write(email)

    return email

# Main Menu

def main():

    while True:

        print("\n------EMAIL GENERATOR -------")
        print("1. Generate Email")
        print("2. Exit")

        choice = input("\nEnter your choice: ")

        if choice == "1":

            sender_name = input("Your Name: ")
            recipient = input("Recipient (Name/Role): ")
            purpose = input("Purpose: ")
            tone = input("Tone (Professional/Casual/Formal/Friendly): ")

            print("\nGenerating Email...\n")

            email = generate_email(
                recipient,
                purpose,
                tone,
                sender_name
            )

            print("=" * 60)
            print(email)
            print("=" * 60)

            print("\n Email saved as email_output.txt")

        elif choice == "2":

            print("\nThank you for using Email Generator!")
            break

        else:

            print("\n Invalid choice! Please try again.")

# Run Program

main()
```
