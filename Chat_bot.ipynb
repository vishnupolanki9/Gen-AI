```python
!pip install -q -U google-generativeai
import google.generativeai as genai
from google.colab import userdata

API_KEY = userdata.get("GOOGLE_API_KEY")
genai.configure(api_key=API_KEY)

# KNOWLEDGE BASE (15 FAQs)

knowledge_base = """
Business Name: Campus Cafe

1. What are the canteen timings?
Answer: The canteen is open from 10:00 AM to 9:00 PM.

2. Is the canteen open on Sunday?
Answer: Yes. The canteen is open on Sundays.

3. Where is the canteen located?
Answer: It is located beside Block B near the college library.

4. Do you provide breakfast?
Answer: Yes. Breakfast is available.

5. What lunch items are available?
Answer: Veg Meals, Chicken Biryani, Fried Rice, Noodles, Burgers, Sandwiches and South Indian meals.

6. Are vegetarian meals available?
Answer: Yes. Multiple vegetarian meals are available every day.

7. What payment methods are accepted?
Answer: Cash, UPI.

8. Do you provide home delivery?
Answer: No. Home delivery service is not available.

9. Can students pre-order food?
Answer: Yes. Students can place an order up to 30 minutes before pickup.

10. Do you offer cold drinks?
Answer: Yes. Soft drinks, bottled water, fresh juices and milkshakes are available.

11. Are combo meals available?
Answer: Yes. Burger Combo, Pizza Combo and Student Meal Combo are available.

12. Is drinking water available?
Answer: Yes. Free RO drinking water is available for everyone.

13. Is seating available?
Answer: Yes. The canteen has seating capacity for approximately 120 students.

14. Who should I contact for complaints?
Answer: Contact the canteen manager at the help desk inside the canteen.

15. Can food be customized?
Answer: Minor customizations such as less spicy or no onions can be requested depending on availability.
"""

# SYSTEM INSTRUCTION
system_instruction = f"""
You are the official FAQ chatbot for Campus Cafe.

Your job is to answer questions ONLY using the knowledge base provided below.

Knowledge Base:

{knowledge_base}

Rules:

1. Answer only from the knowledge base.
2. Do not invent or assume any information.
3. If the answer is not available in the knowledge base, politely reply:
   "I'm sorry, I can only answer questions related to Campus Cafe."
4. Keep answers short, clear and polite.
5. Maintain conversation context naturally.
"""

# CREATE MODEL

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction=system_instruction,
    generation_config={
        "temperature": 0.2
    }
)

# START CHAT

chat = model.start_chat(history=[])

# CHATBOT

print("=" * 55)
print("        Welcome to Campus Cafe FAQ Chatbot")
print("=" * 55)
print("Type 'quit' or 'exit' anytime to end the chat.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() in ["quit", "exit"]:
        print("\nBot: Thank you for visiting Campus Cafe. Have a great day!")
        break

    try:
        response = chat.send_message(user_input)
        print("\nBot:", response.text)

    except Exception as e:
        print("\nError:", e)
```
